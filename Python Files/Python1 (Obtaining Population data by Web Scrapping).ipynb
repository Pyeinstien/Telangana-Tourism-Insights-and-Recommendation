```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_districts_of_Telangana'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify)
```

```python
soup.find('table')
```

```python
soup.find_all('table', class_ ="wikitable sortable")
```

```python
soup.find('table', class_ ="wikitable sortable").text.strip()
```

```python
soup.find('th').text.strip()
```

```python
tab = soup.find_all('table', class_ ="wikitable sortable")[0]
```

```python
tfoot = tab.find('tfoot')
if tfoot:
    tfoot.decompose()
```

```python
tab
```

```python
rows = tab.find_all('tr')
rows
```

```python
for row in rows:
    if row.find('th') and 'Total' in row.find('th').text:
        row.decompose()
        break
```

```python
rows
```

```python
header = tab.find_all('th')
```

```python
header
```

```python
tab
```

```python
titles = [title.text.strip() for title in header]
titles
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(columns = titles)
```

```python
tab_row = tab.find_all('tr')
```

```python
for row in tab_row[1:]:
    row_data = (row.find_all('td'))
    tab_data = [d.text.strip() for d in row_data]
    
    length = len(data)
    data.loc[length] = tab_data
#     print(tab_data)
```

```python
data
```

```python
data.set_index('S.No.', inplace = True)
```

```python
data
```

```python
data = data.drop('Map', axis = 1)
```

```python
data
```

```python
data.info()
```

```python
data = data.astype({
    'No.of mandals': 'int32',
    'Density(per km2)': 'int32',
    'Urban (%)': 'float64',
    'Literacy (%)': 'float64',
    'Sex ratio': 'int32',
})
```

```python
data.info()
```

```python
data['Area (km2)'] = data['Area (km2)'].str.replace(',', '').astype(int)
data['Population(2011 census)'] = data['Population(2011 census)'].str.replace(',', '').astype(int)
```
